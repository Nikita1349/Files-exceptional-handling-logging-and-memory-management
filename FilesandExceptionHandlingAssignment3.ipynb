15.How does Python's garbage collection system work

Reference Counting
Cyclic Garbage Collection

4.gc Module: Provides functions to manually control and inspect garbage collection behavior.
5.Object Pooling: Small objects are reused to improve memory efficiency.

18.What is the difference between os.fork() and multiprocessing in Python
->Both os.fork() and the multiprocessing module are used for creating new processes in Python, but they differ significantly in how they work and how they are used.
1. os.fork():
What is it?: os.fork() is a system-level function available in Python's os module. It is used to create a child process by duplicating the current process.

How it works: When os.fork() is called, the current process (the parent) is copied, and a new child process is created. The new process is identical to the parent, but the return value differs:

In the parent process, os.fork() returns the PID (process ID) of the child process.

In the child process, os.fork() returns 0.

Platform: os.fork() is available only on Unix-like systems (Linux, macOS) and not on Windows.

Usage: It is typically used for lower-level process creation and management. It directly creates a process by copying the parent process.

2. multiprocessing Module:
What is it?: The multiprocessing module is a higher-level Python module designed for creating and managing processes. It provides a cleaner and more flexible API for handling parallelism, especially in cross-platform applications.

How it works: The multiprocessing module uses forking or spawn mechanisms depending on the platform (e.g., forking on Unix, spawn on Windows).

On Unix-like systems, it uses fork() under the hood (similar to os.fork()), but it abstracts away the complexity.

On Windows, it uses a spawn method to create a new Python interpreter in the child process.

```python
#5.How would you catch and handle division by zero error in Python
#In Python, you can catch and handle a division by zero error using a try-except block. The specific exception raised for division by zero is the ZeroDivisionError.
try:
    # Attempting to divide by zero
    result = 10 / 0
except ZeroDivisionError as e:
    # Handle the error by printing a custom message
    print("Error: Cannot divide by zero!")
    print(f"Details: {e}")
else:
    # If no exception occurs, this block is executed
    print(f"The result is {result}")
finally:
    # This block is always executed, regardless of an exception
    print("Execution completed.")
```

```python
#6.Write a Python program that logs an error message to a log file when a division by zero exception occurs
#To log an error message to a log file when a division by zero exception occurs, you can use Python's built-in logging module. Here's a Python program that demonstrates this:
import logging

# Set up logging configuration
logging.basicConfig(
    filename='error_log.txt',  # The log file name
    level=logging.ERROR,       # Log level (only ERROR and more severe will be logged)
    format='%(asctime)s - %(levelname)s - %(message)s',  # Format of log messages
)

def divide(a, b):
    try:
        result = a / b
        return result
    except ZeroDivisionError as e:
        # Log the error with details to the log file
        logging.error(f"Attempted to divide {a} by {b}. Error: {e}")
        print("Cannot divide by zero!")
        return None

# Test the function with division by zero
divide(10, 0)
```

```python
#18.How would you implement a basic logging setup that logs to a file with rotation after 1MB
import logging
from logging.handlers import RotatingFileHandler

# Set up a rotating file handler
log_file = "app.log"
max_log_size = 1 * 1024 * 1024  # 1MB in bytes
backup_count = 3  # Keep 3 backup log files

# Create a handler that rotates logs
handler = RotatingFileHandler(log_file, maxBytes=max_log_size, backupCount=backup_count)

# Set the logging format
formatter = logging.Formatter('%(asctime)s - %(levelname)s - %(message)s')
handler.setFormatter(formatter)

# Set up the logger
logger = logging.getLogger()
logger.setLevel(logging.INFO)  # Set the logging level to INFO
logger.addHandler(handler)

# Example log messages
logger.info("This is an info message.")
logger.error("This is an error message.")
logger.warning("This is a warning message.")
```

```python
#21.Write a Python program that reads a file and prints the number of occurrences of a specific word
def count_word_in_file(file_path, target_word):
    try:
        with open(file_path, "r") as file:
            # Initialize a counter for the occurrences
            word_count = 0

            # Read the file line by line
            for line in file:
                # Split the line into words and count the target word occurrences
                word_count += line.lower().split().count(target_word.lower())

            return word_count
    except FileNotFoundError:
        print(f"Error: The file '{file_path}' does not exist.")
        return 0
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
        return 0

# Example usage
file_path = "example.txt"
target_word = "python"

count = count_word_in_file(file_path, target_word)
print(f"The word '{target_word}' occurred {count} times in the file.")
```
